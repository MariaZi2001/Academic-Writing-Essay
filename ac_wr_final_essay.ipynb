# The Construction of National Identity in Trump’s 2025 Inaugural Address

### Introduction

In political discourse, language is a powerful tool for constructing and reinforcing national identity. This analysis applies principles from Critical Discourse Analysis (CDA) to Donald Trump’s 2025 inaugural address to examine how linguistic choices — particularly the use of pronouns and high-frequency lexical items — contribute to shaping group identity.

The essay focuses on two main aspects: the use of in-group and out-group pronouns and the repetition of specific key words that reflect the speech’s thematic focus. In-group pronouns such as *we* and *our* are used extensively to establish solidarity between the speaker and the audience, while out-group references like *they* and *them* are minimized. This pattern supports a populist narrative centered on unity and shared purpose.

Additionally, an analysis of the ten most frequently used words — including *nation*, *America*, *country*, and *American* — reveals a consistent emphasis on national identity and collective belonging. The lexical choices further underline the speech’s core message: a call for unity and renewal.

## Theoretical Framework

This analysis draws on **Critical Discourse Analysis (CDA)**, particularly the work of **Fairclough (2015)** and **van Dijk (2000)**, who examine how political discourse constructs and reinforces social hierarchies and group divisions. One common strategy in such discourse is the creation of a binary opposition between **“us”** (the in-group) and **“them”** (the out-group).

Pronouns are a key linguistic tool in constructing these group identities:

- **In-group pronouns** such as *we*, *us*, and *our* signal unity, shared values, and collective purpose between the speaker and the intended audience.
- **Out-group pronouns** like *they* and *them* often refer to opponents, outsiders, or perceived threats are used to assign blame, express distance or justify exclusion.

In populist rhetoric, this strategy serves to create a strong emotional bond between the speaker and the in-group, while framing the out-group as dangerous, corrupt or incompetent.

Out-groups are frequently referenced not by name but through generalizing and emotionally charged terms such as “radical establishment”or “criminals.” This indirect language allows for broader interpretations and strengthens ideological polarization without requiring factual specificity.

By examining the frequency and function of these pronouns and labels, it becomes possible to identify the underlying narrative structure of the speech. This framework reveals how language shapes political identity, legitimizes authority and mobilizes support through emotional and symbolic appeals.


## Speech Analysis
The text below is scraped from the White House official speech transcript.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.whitehouse.gov/remarks/2025/01/the-inaugural-address/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
text = ' '.join([p.text for p in soup.find_all('p')])
text = text.replace('\n', ' ')
print(text)
```

## Counting In-group vs. Out-group Pronouns
This step uses regular expressions and Counter to count how often each in-group and out-group pronoun appears.

```python
import re
from collections import Counter
words = re.findall(r'\b\w+\b', text.lower())
in_group = ['we', 'us', 'our', 'ourselves']
out_group = ['they', 'them', 'their', 'themselves']
counter = Counter(words)
in_counts = {p: counter[p] for p in in_group}
out_counts = {p: counter[p] for p in out_group}
in_counts, out_counts
```

## Visualizing Pronoun Use

```python
import matplotlib.pyplot as plt
labels = ['we', 'us', 'our', 'ourselves', 'they', 'them', 'their', 'themselves']
values = [in_counts.get(w, 0) for w in ['we', 'us', 'our', 'ourselves']] + [out_counts.get(w, 0) for w in ['they', 'them', 'their', 'themselves']]
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color='darkcyan')
plt.title("Pronoun Frequency in Trump's Inaugural Address (2025)")
plt.ylabel("Count")
plt.xlabel("Pronoun")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

This aligns with populist rhetorical strategies, which aim to strengthen a shared sense of belonging among "the people."

## Conclusion

Donald Trump’s 2025 inaugural address exhibits a strong reliance on in-group pronouns, particularly **we** (88 instances) and **our** (71 instances). This linguistic choice plays a central role in constructing a collective identity that unites the speaker with the audience. By framing the nation as a single entity with shared goals and values, the address fosters a sense of solidarity and inclusion.

In contrast, out-group pronouns such as **they** (10 instances) and **them** (1 instance) are used sparingly. This limited usage indicates that references to external groups or opponents serve a secondary rhetorical function: rather than attacking others, they help reinforce the central narrative of unity and national cohesion. Such a pattern is characteristic of populist political discourse, which often emphasizes “the people” as a unified collective while minimizing attention to external adversaries.

The lexical analysis of the ten most frequently used words further supports this interpretation:

The frequent use of emotionally charged and nationalistic terms reinforces the address’s core rhetorical goal: to unify the audience under a shared vision of patriotic renewal and collective purpose.

Overall, the combination of pronoun patterns and lexical frequency demonstrates how political language is strategically employed to construct unity, evoke emotional resonance and define the speaker’s alignment with “the people.”
